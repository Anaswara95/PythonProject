### Find out under which team and position, spending in terms of salary is high.

```python
x = df['Salary'].mean()
df['Salary'].fillna(x,inplace = True)
df
```

```python
df1=df[['Team','Position','Salary']]
df1
```

```python
total_salary = df1.groupby(['Team', 'Position'])['Salary'].sum()
max_spending = total_salary.idxmax()
max_spending_salary = total_salary.max()
print(f"The team and position with the highest spending is:\n Team: {max_spending[0]}\n position: {max_spending[1]} \n with a total salary of {max_spending_salary}.")
```

```python
plt.figure(figsize=(20,5))
sns.scatterplot(y="Salary", x="Team", hue="Position", data=df1)
plt.ylabel("Salary")
plt.xlabel("Team")
plt.title("Scatter Plot of Position vs Team with high Salary")
plt.xticks(rotation=45)
plt.show()
```

### Data Story:
- The scatter plot presents the salary distribution across different positions within each team in the dataset. It becomes evident that the Los Angeles Lakers stand out as the highest-spending team.
- The Los Angeles Lakers spend the most, notably paying 31,866,445 to **SF**. This shows they prioritize investing heavily in their SF players compared to other positions on their team..

### Find if there is any correlation between age and salary , represent it visually.

```python
df2=df[['Age','Salary']]
df2
```

```python
corr = df2.corr()
print(corr)
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Correlation Matrix between Age and Salary of Employee Dataset")
plt.show()
```

### Data Story:
The correlation shows the relationship between "Age" and "Salary" in the employee dataset:

- **Age vs. Age (1.00000)** : This value of 1.00000 along the diagonal represents the correlation of a variable with itself, which is always 1. It indicates a perfect positive correlation, as expected.

- **Age vs. Salary (0.21117)** : The correlation coefficient of 0.21117 suggests a positive but relatively weak relationship between age and salary. This means that as age increases, there's a tendency for salaries to increase slightly, but the correlation isn't very strong.

- Similarly, **Salary vs. Age (0.21117)** : Since correlation is a symmetric measure, this value mirrors the correlation between "Age" and "Salary" and reaffirms the weak positive relationship between the two variables.

In simpler terms, the correlation matrix indicates that there's a positive but not particularly strong relationship between age and salary in the employee dataset. As employees' ages tend to increase, there is a tendency for their salaries to slightly rise.
